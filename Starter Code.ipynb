# 4.0 Example Visualizations

Let's make some visualizations!
Below are some examples of visualizations that you can make using `matplotlib`, `seaborn`, `altair`, and `plotly`.

## 4.1 Matplotlib
Here's how you create a bar chart for a single column using `matplotlib`.

```python
# Get the top 10 committees by number of contributions
top_committees = sample_contribution4_2020['com_legal_name'].value_counts().head(10)

# Create a bar chart
plt.figure(figsize=(10, 8)) # Increased figure height
plt.bar(top_committees.index, top_committees.values)
plt.xlabel('Committee Name')
plt.ylabel('Number of Contributions')
plt.title('Top 10 Committees by Number of Contributions in 2020')
plt.xticks(rotation=90)
plt.show()
```

## 4.2 Seaborn

Here's how you can create a scatter plot for two columns using `seaborn`.

```python
# Clean and convert amount + aggregate to numbers
for col in ["amount", "aggregate"]:
    sample_contribution4_2020[col] = (
        sample_contribution4_2020[col]
        .astype(str)
        .str.replace(",", "", regex=False)
        .str.replace("$", "", regex=False)
        .replace({"": None, " ": None})
        .astype(float)
    )

# Create a scatter plot
sns.scatterplot(
    data=sample_contribution4_2020,
    x="amount",
    y="aggregate",
    alpha=0.5  # Adjust to make points semi-transparent for better visualization
)
plt.title("Scatterplot of Amount vs. Aggregate")
plt.xlabel("Amount ($)")
plt.ylabel("Aggregate ($)")
plt.show()
```

## 4.3 Altair

Here's how you can create a box plot for a single column using `altair`.

```python
df = sample_contribution4_2020.copy()
df["amount"] = pd.to_numeric(df["amount"], errors="coerce")
df = df.dropna(subset=["amount", "com_type"])

# Sum Total per Committee Type
df_summary = df.groupby("com_type", as_index=False)["amount"].sum()

# Disable Altair's 5k Max Rows
alt.data_transformers.disable_max_rows()

chart = (
    alt.Chart(df_summary)
    .mark_bar()
    .encode(
        x=alt.X("amount:Q", title="Total Contribution Amount ($)"),
        y=alt.Y("com_type:N", sort="-x", title="Committee Type"),
        tooltip=["com_type", alt.Tooltip("amount:Q", format=",.0f")]
    )
    .properties(
        title="Total Contributions by Committee Type (2020)"
    )
)

chart.show()
```

## 4.4 Plotly

Here's how you can a box plot using `plotly`.
Note that you get built-in interactivity with Plotly (e.g., hovertips, built-in zooming)!

```python
fig = px.box(
    df,
    x="amount",
    y="com_type",
    title="Distribution of Contribution Amounts by Committee Type (2020)",
    labels={"amount": "Contribution Amount ($) in Log Scale", "com_type": "Committee Type"}
)

# Use a Log Scale to show lower contributions better
fig.update_xaxes(type="log")

fig.show()
```
